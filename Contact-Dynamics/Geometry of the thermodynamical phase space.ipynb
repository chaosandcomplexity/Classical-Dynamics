# Geometry of the thermodynamical phase space

## Introduction

Several areas have been benefited from the use of geometrical tools, gravitation with the Riemannian geometry, classical mechanics with the use of symplectic structures, or the particle physics with the use of convex polytopes in the calculation of scattering amplitudes. It is natural to expect that other areas could be benefited from the use of geometrical methods. In particular, thermodynamics and statistical mechanics enjoy this kind of characterization in terms of a geometrical structure widely know in mathematics, the so-called contact structures. This kind of structure corresponds to an odd dimensional analogous of the symplectic structures, structures well known in physics due to his use in classical mechanics in the study of chatotic behaviour and semiclassical aproximations of quantum mechanics.

The symplectic case is of great interest due to his description of the classical phase space $M$. For example, the classical dynamics described by a Hamiltonian function $H(p,q;t)$ can be understood in terms of integral curves of a vector field $X_H$ (curves that have as tangent vectors the vector field $X_H$) defined by the function $H$ and by the symplectic structure of the space. By studying the geometrical properties of $X_H$ we can arrive to the Hamilton's equations of motion from the components of $X_H$ in the local coordinates $(p,q)$ of the symplectic manifold $M$, undestanding the corresponding integral curves as curves where the function $H(p,q)$ (without time dependence) takes constant values. Finally we can make the statement that a mechanical system is fully characterized by it's Hamiltonian function $H$.

Taking as reference the classical phase space with it's symplectic formalism is natural to expect that the thermodynamics to benefic from being treated in the same footing. This is possible by using a thermodynamical phase space $(p,q,w)$ equiped with a contact structuture.

## Contact geometry essentials

Let $\mathcal{C}$ be a $(2n+1)$ dimensional manifold. It is said that the pair $(\mathcal{C},\lambda)$ with $\lambda$ a 1-form form a contact manifold. $\lambda$ is such that

\begin{equation*}
\lambda \wedge (d\lambda)^n \neq 0,
\end{equation*}

where $\wedge$ denotes the wedge product between diferential forms. This condition can be understood as the requirement of having a non vanishing volumen element defined over the manifold.

## Dynamics from a master equation

### Markovian case

Be $\Gamma$ a finite set of discrete states, $t \in \mathbb{R}$ time, and $\rho(j,t)dt$ the probability of having the state $j \in \Gamma$  between a time $t$ and $t + dt$ with $\rho: \Gamma \times \mathbb{R} \rightarrow \mathbb{R}_{\geq 0}$ a time dependent probability distribution. Let be  $\rho_\theta^{eq}(j)$ a distribution that can be written in the form

\begin{equation*}
	\rho_\theta^{eq}(j) = \frac{\pi_\theta(j)}{\mathcal{Z(\theta)}},
\end{equation*}

where $\theta \in \Theta \subset \mathbb{R}^n$ is a parameter set, and $\mathcal{Z}: \Theta \rightarrow \mathbb{R}$ is the so called partition function defined such that  $\rho_\theta^{eq}$ be normalized

\begin{equation*}
	\mathcal{Z}(\theta) = \sum_{j\in \Gamma} \pi_\theta(j).
\end{equation*}

Consider the set of master equations with Markovian kernel

\begin{equation*}
	\frac{\partial }{\partial t} \rho(j,t) = \sum_{j' (\neq j)} [w(j|j') \rho(j',t)  -  w(j'|j)\rho(j,t)],
\end{equation*}

where $w(j|j')$ denotes the probability of jumping from a state $j$ to $j'$. Choosing $w_\theta(j|j') = p_\theta^{eq}$ with $p_\theta^{eq}(j) \neq 0$ for all $j \in \Gamma$ we arrive to

\begin{equation}
\frac{\partial }{\partial t} \rho(j,t) = \sum_{j' (\neq j)} [w_\theta(j|j') \rho(j',t)  -  w_\theta(j'|j)\rho(j,t)] = \rho_\theta^{eq}(j) - \rho(j,t).
\end{equation}

The solution to this equation is given  by

\begin{equation*}
	\rho(j,t) = e^{-t}\rho(j,0) + (1 - e^{-t})\rho_\theta^{eq}(j),
\end{equation*}

expression that allows us to conclude immediately 

\begin{equation*}
	\lim_{t \to \infty} \rho(j,t) = \rho_\theta^{eq}.
\end{equation*}

Let $\rho_\theta^{eq}$ denote a family of distributions such that $\pi_\theta(j) = \text{exp}(\theta^a \mathcal{O}_a(j))$ with $\{\mathcal{O_a}\}$ some function set. This family parametrized by $\theta = \{\theta^1,...,\theta^n\} \in \Theta \subset \mathbb{R}^n$ can be written in the following form

\begin{equation*}
	\rho_\theta^{eq}(j) = \text{exp}(\theta^a \mathcal{O}_a - \Psi_{eq}(\theta)),
\end{equation*}

where $\Psi(\theta)$ is such that $	\rho_\theta^{eq}(j)$ is normalized

\begin{equation*}
	\Psi_{eq}(\theta) := ln\Bigg( \sum_{j \in \Gamma} e^{\theta^a \mathcal{O}_a(j)} \Bigg).
\end{equation*}

Through a function $\Psi(\theta)$ defined for some distribution of this family that not necessarily describes the thermodynamical equilibrium of some system is possible to define the notion of conjugate thermodynamical variables. The quantities $\theta^a$ and $\langle \mathcal{O}_a \rangle$ are said to define that pair in the following sense

\begin{equation*}
	\eta_a := \langle \mathcal{O}_a \rangle = \frac{\partial \Psi}{\partial \theta^a}, \hspace{0.8 cm} a \in \{1,...,n\}.
\end{equation*}

Now we define the expected values with respect to $\rho(j,t;\theta)$ and $\rho_\theta^{eq}(j)$ as

\begin{equation*}
	\langle \mathcal{O}_a \rangle_\theta (t) := \sum_{j \in \Gamma} \mathcal{O}_a (j) \rho(j,t;\theta) \hspace{0.3 cm} \text{y} \hspace{0.3 cm} \langle \mathcal{O}_a \rangle^{eq} := \sum_{j \in \Gamma} \mathcal{O}_a (j) \rho_\theta^{eq}(j)
\end{equation*}

respectivaly. The definition of $\Psi$ can be extended for non equilibrium states in the following way

\begin{equation*}
	\Psi(\theta,t) := \Bigg( \frac{1}{J_0} \sum_{j \in \Gamma} \frac{\rho(j,t;\theta)}{\rho_\theta^{eq}(j)} \Bigg) \Psi_{eq}(\theta), \hspace{0.3 cm} \text{with} \hspace{0.3 cm} J_0 := \sum_{j' \in \Gamma} 1.
\end{equation*}

Finally, using the master equation and the definitions for the expected values it can be found that the dynamics that follow the conjugate thermodynamical quantities $\theta_a$ y $\langle \mathcal{O}_a \rangle$ and the function $\Psi$ asocieted to $\rho(j,t;\theta)$ is given by the following set of equations

\begin{equation}
	\frac{d}{dt} \theta^a = 0, \hspace{0.3 cm} \frac{d}{dt} \langle \mathcal{O}_a \rangle_\theta = - \langle \mathcal{O}_a \rangle_\theta + \frac{\partial \Psi_{eq}}{\partial \theta^a}, \hspace{0.3cm} \frac{d}{dt} \Psi = -\Psi + \Psi_{eq}
\end{equation}
